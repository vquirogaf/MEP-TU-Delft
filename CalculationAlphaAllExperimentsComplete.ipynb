```python
import numpy as np

def derivative_linear2(df, steps):
    # Extract data from the DataFrame
    intensity = df['Mean_Intensity_norm_max_mean'].iloc[1:]
    receptors = df['recept_'].iloc[1:]

    # Calculate natural logarithms once
    log_intensity = np.log(intensity)
    log_receptors = np.log(receptors)

    deriv2 = np.diff(log_intensity, n=steps) / np.diff(log_receptors, n=steps)
    xval2 = receptors[:-steps]
    theta2 = intensity[:-steps]

    return xval2, deriv2, theta2
```
